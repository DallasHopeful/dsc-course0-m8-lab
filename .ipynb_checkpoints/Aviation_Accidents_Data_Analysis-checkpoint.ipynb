**Distribution of injury rates: small makes**

Use a violinplot to look at the distribution of the fraction of passengers serious/fatally injured for small airplane makes. Just display makes with the ten lowest mean serious/fatal injury rates.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Check current directory
print(f"Current working directory: {os.getcwd()}")

# Load data
Clean_Aviation = pd.read_csv('data/AviationDataClean.csv', low_memory=False)

# Create a new column 'Aircraft_Size' based on total_passengers
Clean_Aviation['Aircraft_Size'] = Clean_Aviation['total_passengers'].apply(lambda x: 'large' if x >= 20 else 'small')

# Calculate Fatal/Serious Injury Fraction
Clean_Aviation['Fatal_Serious_Injury_Fraction'] = (
    (Clean_Aviation['Total.Fatal.Injuries'] + Clean_Aviation['Total.Serious.Injuries']) /
    Clean_Aviation['total_passengers']
)

# Replace infinite values and NaNs resulting from division by zero with 0
Clean_Aviation['Fatal_Serious_Injury_Fraction'].replace([float('inf'), -float('inf')], 0, inplace=True)
Clean_Aviation['Fatal_Serious_Injury_Fraction'].fillna(0, inplace=True)

# Filter data for small aircraft size
small_planes = Clean_Aviation[Clean_Aviation['Aircraft_Size'] == 'small']

# Calculate mean injury fraction per Make_Model_ID
mean_injury_fraction = small_planes.groupby('Make_Model_ID')['Fatal_Serious_Injury_Fraction'].mean().reset_index()

# Sort to find 10 lowest injury fractions
lowest_10_models = mean_injury_fraction.sort_values(by='Fatal_Serious_Injury_Fraction').head(10)

# Filter original data to include only these 10 Make_Model_IDs
lowest_10_data = small_planes[small_planes['Make_Model_ID'].isin(lowest_10_models['Make_Model_ID'])]

# Plot violin plot
plt.figure(figsize=(14, 8))
sns.violinplot(
    x='Fatal_Serious_Injury_Fraction', 
    y='Make_Model_ID', 
    data=lowest_10_data, 
    inner='quartile', 
    scale='width'
)

plt.title('10 Small Aircraft Models with Lowest Fatal/Serious Injury Fraction')
plt.xlabel('Fatal/Serious Injury Fraction')
plt.ylabel('Aircraft Make and Model')
plt.tight_layout()
plt.show()
```

**Distribution of injury rates: large makes**

Use a stripplot to look at the distribution of the fraction of passengers serious/fatally injured for large airplane makes. Just display makes with the ten lowest mean serious/fatal injury rates.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

print(f"Current working directory: {os.getcwd()}")

Clean_Aviation = pd.read_csv('data/AviationDataClean.csv', low_memory=False)

# Create a new column 'Aircraft_Size' based on total_passengers
Clean_Aviation['Aircraft_Size'] = Clean_Aviation['total_passengers'].apply(lambda x: 'large' if x >= 20 else 'small')

# Calculate Fatal/Serious Injury Fraction
Clean_Aviation['Fatal_Serious_Injury_Fraction'] = (
    (Clean_Aviation['Total.Fatal.Injuries'] + Clean_Aviation['Total.Serious.Injuries']) /
    Clean_Aviation['total_passengers']
)

# Replace infinite values and NaNs resulting from division by zero with 0
Clean_Aviation['Fatal_Serious_Injury_Fraction'].replace([float('inf'), -float('inf')], 0, inplace=True)
Clean_Aviation['Fatal_Serious_Injury_Fraction'].fillna(0, inplace=True)

# Filter data to only large aircraft
large_aircraft = Clean_Aviation[Clean_Aviation['Aircraft_Size'] == 'large']

# Calculate mean serious/fatal injury fraction by make
mean_injury_by_make = large_aircraft.groupby('Make')['Fatal_Serious_Injury_Fraction'].mean().reset_index()

# Get the ten makes with the lowest mean injury fraction
lowest_ten_makes = mean_injury_by_make.nsmallest(10, 'Fatal_Serious_Injury_Fraction')['Make']

# Filter the original data to only include these ten makes
lowest_ten_data = large_aircraft[large_aircraft['Make'].isin(lowest_ten_makes)]

# Plot using seaborn stripplot
plt.figure(figsize=(12, 8))
sns.stripplot(data=lowest_ten_data, 
              x='Fatal_Serious_Injury_Fraction', 
              y='Make', 
              jitter=True, 
              alpha=0.6)

plt.title('Distribution of Fatal/Serious Injury Fraction for Large Aircraft (10 Lowest Mean Makes)')
plt.xlabel('Fatal/Serious Injury Fraction')
plt.ylabel('Aircraft Make')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Evaluate the rate of aircraft destruction for both small and large aircraft by Make.** 

Sort your results and keep the lowest 15.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

print(f"Current working directory: {os.getcwd()}")

Clean_Aviation = pd.read_csv('data/AviationDataClean.csv', low_memory=False)

# Create a new column 'Aircraft_Size' based on total_passengers
Clean_Aviation['Aircraft_Size'] = Clean_Aviation['total_passengers'].apply(lambda x: 'large' if x >= 20 else 'small')

# Calculate Fatal/Serious Injury Fraction
Clean_Aviation['Fatal_Serious_Injury_Fraction'] = (
    (Clean_Aviation['Total.Fatal.Injuries'] + Clean_Aviation['Total.Serious.Injuries']) /
    Clean_Aviation['total_passengers']
)

# Replace infinite values and NaNs resulting from division by zero with 0
Clean_Aviation['Fatal_Serious_Injury_Fraction'].replace([float('inf'), -float('inf')], 0, inplace=True)
Clean_Aviation['Fatal_Serious_Injury_Fraction'].fillna(0, inplace=True)

# Calculate the destruction rate by aircraft size and make
destruction_rate = Clean_Aviation.groupby(['Make', 'Aircraft_Size'])['Aircraft.damage'].apply(
    lambda x: (x == 'Destroyed').mean()
).reset_index(name='Destruction_Rate')

# Sort by destruction rate ascending and keep the lowest 15
lowest_15_destruction_rate = destruction_rate.sort_values(by='Destruction_Rate', ascending=True).head(15)

# Display the result
print(lowest_15_destruction_rate)

# Visualize the results
plt.figure(figsize=(12, 8))
sns.barplot(
    data=lowest_15_destruction_rate,
    x='Destruction_Rate',
    y='Make',
    hue='Aircraft_Size',
    palette='viridis'
)
plt.xlabel('Destruction Rate')
plt.ylabel('Aircraft Make')
plt.title('Lowest 15 Aircraft Makes by Destruction Rate (Small vs Large Aircraft)')
plt.legend(title='Aircraft Size')
plt.tight_layout()
plt.show()
```
